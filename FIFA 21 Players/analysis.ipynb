### Data Cleaning

```python
# Dropping unnecessary columns

df.drop(['photoUrl', 'playerUrl', 'Loan Date End'], axis=1, inplace=True)
```

#### Converting the value of Height and Weight into cm and kg

```python
df['Height']
```

```python
df['Height'] = pd.Series(df['Height']) # Converting to Series

# Converting height to cm

def height_to_cm(x):
	h = x.split("'") # Splitting by '
	return int(h[0]) * 30.48 + int(h[1].rstrip('"')) * 2.54 # Converting to cm

df['Height'] = df['Height'].apply(height_to_cm) # Applying the function
```

```python
df['Height'] = df['Height'].astype('int64') # Converting to int64
df['Height']
```

```python
df['Weight']
```

```python
df['Weight'] = pd.Series(df['Weight'])

def weight_to_kg(x):
    s = str(x)
    s = s.replace('lbs', '')
    return int(s)/2.205

df['Weight'] = df['Weight'].apply(weight_to_kg)
```

```python
df['Weight'] = df['Weight'].astype('int64')
df['Weight']
```
